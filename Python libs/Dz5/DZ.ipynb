Задание 1.

```python
import numpy as np
import matplotlib.pyplot as plt

def complex_function(x):
    return np.sin(x) + np.cos(2*x) * np.tan(2*x) / (1 + x**2)

# Генерация X-координат точек
x_values = np.linspace(-10, 10, 1000)

# Вычисление Y-значений с использованием созданной функции
y_values = complex_function(x_values)

# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, linestyle='dashed', color='red', alpha=0.7, label='Вот такая моя функция')

# Добавление подписей и заголовка
plt.xlabel('X-координата')
plt.ylabel('Y-координата')
plt.title('График алгебраической функции')
plt.legend()

# Отображение графика
plt.grid(True)
plt.show()
```

Задание 2.

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация значений X и Y
X = np.random.normal(0, 1, 3000)
Y = np.random.normal(3, 4, 3000)

# Построение графика
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, marker='<', color='green', alpha=0.4, s=40)

# Добавление подписей и заголовка
plt.xlabel('X-координата')
plt.ylabel('Y-координата')
plt.title('Распределение точек')

# Отображение графика
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация данных результатов забега школьников на 100м.
data = np.random.normal(16, 2, 1000)

# Построение гистограммы
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='red', alpha=0.5, edgecolor='black')

# Добавление подписей и заголовка
plt.xlabel('Время (секунды)')
plt.ylabel('Частота')
plt.title('Гистограмма результатов забега школьников на 100 метров')

# Отображение гистограммы
plt.show()
```
