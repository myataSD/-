```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
X = np.array([1, 3, 7])
y = np.array([2, 6, 14])

# Инициализация весов
w = 0
b = 0

# Коэффициент обучения
lr = 0.01

# Количество итераций
num_iterations = 3500

# Обучение модели
for i in range(num_iterations):
    # Прямой проход
    y_pred = w * X + b

    # Обратный проход
    error = y - y_pred
    w_grad = -2 * np.mean(X * error)
    b_grad = -2 * np.mean(error)

    # Обновление весов
    w -= lr * w_grad
    b -= lr * b_grad

# Вывод полученных весов
print("Полученные веса:")
print("w:", w)
print("b:", b)

# Построение графика прямой
plt.scatter(X, y)
plt.plot(X, w * X + b, color='red')
plt.show()

# Вычисление настоящих весов
w_real = (np.mean(X * y) - np.mean(X) * np.mean(y)) / (np.mean(X**2) - np.mean(X)**2)
b_real = np.mean(y) - w_real * np.mean(X)

# Сравнение полученных весов с настоящими
print("Настоящие веса:")
print("w:", w_real)
print("b:", b_real)

# Проверка схожести весов
print("Схожи ли полученные веса с настоящими?")
print(np.allclose(w, w_real) and np.allclose(b, b_real))
```
